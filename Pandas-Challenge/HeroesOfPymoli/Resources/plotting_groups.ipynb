```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
used_string = "../Resources/used_cars.csv"
used_car_df = pd.read_csv(used_string)

used_car_df.head()
```

```python
#using groupby function 
#returns the count as a series

maker_group = used_car_df.groupby("maker")
count_makers = maker_group["maker"].count()
count_makers
```

```python
count_chart = count_makers.plot(kind="bar", facecolor="red")

count_chart.set_xlabel("Car Manufactures")
count_chart.set_ylabel("Number of Cars")

plt.show()
plt.tight_layout()
```
