```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data = pd.read_csv("Resources/got.csv")
got_data.head()
```

```python
# Get attacker and defender data
attacker_data = got_data["attacker_king"].value_counts()
defender_data = got_data["defender_king"].value_counts()
```

```python
#This shows 2 kings defended but never attacked - you can almost assume that there are 2 blank values you need to fill
print(attacker_data.count())
print(defender_data.count())
```

```python
# Get total battle data - you can add defender data to the attacker data since you used count function above
#to use the add function you need the empty values to be a 0 so it is added correctly (fill_value=0) or any value of your choosing
battle_data = attacker_data.add(defender_data, fill_value=0)
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar", facecolor="red")
```

```python
# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
plt.xlabel("Kings")
```

```python
# Show plot
plt.show()
plt.tight_layout()
```
